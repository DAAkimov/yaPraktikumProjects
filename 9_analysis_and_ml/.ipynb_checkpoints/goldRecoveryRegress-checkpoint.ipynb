Проверим, что в полученных данны значение `rougher.output.recovery` посчитано правильно.

Для этого напишем функцию рассчета значения recovery и вызовем её для всех имеющихся значений.

```python
def recovery(row, stage):
    C = row[stage + '_output_concentrate_au']
    F = row['rougher_input_feed_au']
    T = row[stage + '_output_tail_au']
    if F == 0 or C-T == 0:
        return np.NaN
    else:
        return C * (F - T) / (F * (C - T)) * 100 
```

```python
not_nan_rec = ~train_df['rougher_output_recovery'].isna()

manual_recovery = train_df[not_nan_rec].apply(recovery, args=('rougher',), axis=1)
manual_recovery = manual_recovery.dropna()

print(MAE(train_df.loc[manual_recovery.index,'rougher_output_recovery'], manual_recovery))
print("Not NaN recovery list len:", len(not_nan_rec))
print("Manual recovery list len:", len(manual_recovery))
```

Ошибка почти нулевая, следовательно полученные данные содержат правильные расчеты. Мы могли бы расчитать недостающие данные для признака `rougher.output.recovery`, но мы видим, что данных для восстановления этого значения меньше, чем собранных.

## 1.3 Предобработка данных.
Первом делом отфильтруем значения. Отбросим по 0.01 квантиль сверху и снизу признака. Так же необходимо учесть, что в некоторых наблюдениях мы имеем близкие к 0 входные значения, которые просто распространяют этот 0 на все дальнейшие вычисления.

Для того, чтобы иметь возможность адекватно обрабатывать пропуски нарисуем матрицу корреляции для всех наших признаков.

```python
plt.figure(figsize=(20,20))
sns.heatmap(rougher_train_df.dropna().corr(method='spearman'), annot=True)
```

```python
def filtr(df):
    new_df = df.copy()
    for columns in df.columns:
        filtr = ((new_df[columns] > new_df[columns].quantile(0.01)) &
                 (new_df[columns] < new_df[columns].quantile(0.99)))
        new_df.loc[:,columns] = new_df.loc[filtr, columns]
    
    return new_df
```

```python
rougher_train_df = rougher_train_df.query('rougher_input_feed_au > 0.1')
rougher_train_df_f = filtr(rougher_train_df)

primary_train_df_f = filtr(primary_train_df)
secondary_train_df_f = filtr(secondary_train_df)
final_train_df_f = filtr(final_train_df)

print("rougher shape:", rougher_train_df_f.shape)
rougher_train_df.head(5)
```

Очистим наши данные от строк с большим количеством пропусков в важных для нас параметрах: входных и выходных характеристиках. Оставшиеся пропуски заполним ближайшими значениями, так как все эти процессы протекают по времени и имеют схожие характеристики для ближайших наблюдений.

```python
important_col = [x for x in rougher_train_df_f.columns if ('_feed_' in x or 'output' in x)]
print(len(important_col))
```

```python
rougher_train_df_f = rougher_train_df_f[rougher_train_df_f[important_col].isna().sum(axis=1)<6]
rougher_train_df_f.shape
```

Мы удалили неинформативные строки, строки содержащие много пропусков и очевидные выбросы. Удалим строки, имеющие пропуски в признаках участвующих в расчете значений Recovery.

Так как признаки floatbank10 имею близкие значения по времени, то мы можем заполнить пропуски с помощью ближайших значений. Зная, что признаки floatbank10 и floatbank11 очень похожи заменим недостающие значения на существующие.

```python
def floatbank_fill(df, point_fltbnk):
    floatbank_col = [x for x in df.columns if (point_fltbnk in x)]
    df[floatbank_col] = df[floatbank_col].fillna(method='ffill')              
```

```python
col = ['rougher_input_feed_au', 'rougher_input_feed_size', 
       'rougher_output_concentrate_au', 'rougher_output_tail_au']
rougher_train_df_f = rougher_train_df_f.dropna(subset=col)

floatbank_fill(rougher_train_df_f, 'floatbank10')
floatbank_fill(primary_train_df_f, 'floatbank')
floatbank_fill(secondary_train_df_f, 'floatbank')

filtr = rougher_train_df_f['rougher_input_floatbank11_sulfate'].isna()
rougher_train_df_f.loc[filtr, 'rougher_input_floatbank11_sulfate'] = rougher_train_df_f['rougher_input_floatbank10_sulfate']

filtr = rougher_train_df_f['rougher_input_floatbank11_xanthate'].isna()
rougher_train_df_f.loc[filtr, 'rougher_input_floatbank11_xanthate'] = rougher_train_df_f['rougher_input_floatbank10_xanthate']
```

```python
primary_train_df_f = primary_train_df_f.loc[rougher_train_df_f.index, :]
secondary_train_df_f = secondary_train_df_f.loc[rougher_train_df_f.index, :]
final_train_df_f = final_train_df_f.loc[rougher_train_df_f.index, :]
```

```python
rougher_train_df_f.shape
```

## <a id="3"> </a>3. Построение модели.

### <a id="3.1"> </a>3.1 Подготовка к построению модели.
Перед построением модели необходимо написать функцию рассчета метрики качества sMAPE.

```python
def smape(target, predict):
    return np.mean((np.abs(target - predict) / ( (np.abs(target) + np.abs(predict)) / 2 ))) * 100

def final_smape(smape_rough, smape_final):
    return 0.25*smape_rough + 0.75*smape_final
```

Теперь мы можем заняться подбром моделей. Так как мы решили строить две разных модели, то сформируем две разные выборки.

```python
def df_construct(df):
    feature_col = [x for x in df.columns if ('input' in x or 'state' in x or 'recovery' in x)]
    target_col = [x for x in df.columns if ('recovery' in x)]
    
    _feature = df[feature_col].dropna()
    _target = _feature[target_col]
    
    return _feature.drop(target_col, axis=1), _target
```

```python
rougher_input, rougher_target = df_construct(rougher_train_df_f)

final_input = pd.concat([primary_train_df_f, secondary_train_df_f, final_train_df_f], axis=1)

cleaner_input, cleaner_target = df_construct(final_input)
```

```python
X_first_train, X_first_val, y_first_train, y_first_val =  train_test_split(rougher_input, rougher_target,
                                                           test_size=0.25, random_state=rand)

X_final_train, X_final_val, y_final_train, y_final_val =  train_test_split(cleaner_input, cleaner_target,
                                                           test_size=0.25, random_state=rand)
```

Теперь сформируем тестовую выборку, не забудем, что мы решили отфильтровать околонулевые значения. Целевой параметр подсмотрим в полной выборке.

```python
test_df_full = test_df.merge(full_df[['date', 'rougher.output.recovery', 'final.output.recovery']],
                             on='date',
                             how='left',
                            )

X_first_test = test_df_full.query('rougher_input_feed_pb + \
                       rougher_input_feed_au + \
                       rougher_input_feed_ag + \
                       rougher_input_feed_sol  > 1 and primary_cleaner_input_feed_size > 1')

rougher_col = [x for x in X_first_test if ('rougher' in x)]
rough_test_input, rough_test_target = df_construct(X_first_test[rougher_col])

not_rougher_col = [x for x in X_first_test if ('rougher' not in x)]

final_test_input, final_test_target = df_construct(X_first_test[not_rougher_col])
```

Теперь мы получили два набора данных, которые можем подавать на вход наших моделей. Изучим 3 модели:
    - Линейную регрессию;
    - Случайные лес;
    - Методо k-ближайших соседей.

### <a id="3.2"> </a> 3.2 Линейная регрессия.
Линейный регрессор самый простой из этих методов, он не требует обучения или подгонки параметров и однозначно рассчитывается по имеющимся признакам.

```python
model_1 = LinearRegression()
model_2 = LinearRegression()

model_1.fit(X_first_train, y_first_train)
model_2.fit(X_final_train, y_final_train)

pred_1 = model_1.predict(X_first_val)
pred_2 = model_2.predict(X_final_val)

smape_rought = smape(y_first_val.values.ravel(), pred_1)
smape_final = smape(y_final_val.values.ravel(), pred_2)

print("First stage sMAPE: {:.2f}%".format(smape_rought))
print("Final stage sMAPE: {:.2f}%".format(smape_final))

print('Final sMAPE: {:.2f}%'.format(final_smape(smape_rought, smape_final)))
```

Линейная регрессия показывает очень хорошее значение sMAPE, это может быть связано с тем, что мы почистили выбросы в данных.



### <a id="3.3"> </a> 3.3 Случайный лес.
Попробуем обучить модель и подобрать оптимальные гиперпараметры. Для оценки модели будем использовать кросс-валидацию, поэтому будем использовать всю тренировочную выборку. Обернем нашу функцию подсчета sMAPE для использования ее как метрики в алгоритме.

```python
my_scorer = make_scorer(smape)
```

```python
def model_fit(model_rough, model_clean, rough_features, rough_target, clean_features, clean_target):
    cv_score_rought = cross_val_score(model_rough, X=rough_features,
                                          y=rough_target.values.ravel(), scoring=my_scorer,
                                          cv=5, n_jobs=-1)
    cv_score_clean = cross_val_score(model_clean, X=clean_features,
                                         y=clean_target.values.ravel(), scoring=my_scorer,
                                         cv=5, n_jobs=-1)

    
    cv_final_mean = final_smape(cv_score_rought.mean(), cv_score_clean.mean())
    
    return [model_rough, model_clean], cv_final_mean
```

```python
n_depth = [3, 6, 8, 12]
n_estimators = [3, 5, 10, 20, 50]

teatch_info = list()

best_sMAPE_score = 1000
forest_models = dict()

for depth in n_depth:
    for estimator in n_estimators:
        model_f = RandomForestRegressor(n_estimators=estimator, max_depth=depth, random_state=rand)
        model_c = RandomForestRegressor(n_estimators=estimator, max_depth=depth, random_state=rand)
        
        model, cv_mean_score = model_fit(model_f, model_c,
                                         rougher_input, rougher_target,
                                         cleaner_input, cleaner_target)
        
        if cv_mean_score < best_sMAPE_score:
            best_sMAPE_score = cv_mean_score
            forest_models['_'.join([str(depth), str(estimator)])] = model
        
        print("Estimator: {:.0f}, Depth: {:.0f}, CV_sMAPE_score: {:.2f}%".format(estimator,
                                                                                depth,
                                                                                cv_mean_score))
        

```

###  <a id="3.4"> </a> 3.4 Методо k-ближайших соседей.
Для данного метода необходимо масштабировать наши признаки.

```python
scaler_1 = MinMaxScaler()
scaler_2 = MinMaxScaler()

x_rough_scaler = scaler_1.fit_transform(rougher_input)
x_clean_scaler = scaler_2.fit_transform(cleaner_input)
```

```python
neighbor_model = dict()
best_sMAPE_score = 100

K_s= [5, 10, 20, 30, 50, 100, 150, 200, 500]

for K in K_s:
    model_r = KNeighborsRegressor(n_neighbors=K)
    model_c = KNeighborsRegressor(n_neighbors=K)
    
    model, cv_score = model_fit(model_r, model_c,
                                x_rough_scaler, rougher_target,
                                x_clean_scaler, cleaner_target)

    if cv_score < best_sMAPE_score:
        best_sMAPE_score = cv_score
        neighbor_model[str(K)] = model
        
    print("neighbors_K: {:.0f}, CV_sMAPE_score: {:.2f}%".format( K, cv_score))
        
```

Метод ближайших соседей показывает наилучший результат по нашим моделям, так как при увеличении параметра n_neighbors свыше 20 наша модель не сильно улучшается, возьмем эту модель за лучшую, чтобы избежать сильного переобучения.

###  <a id="3.5"> </a> 3.5 Тестирование модели.
Теперь проверим нашу модель на тестовой выборке, но для этого нужно её повторно обучить.

```python
best_model_rought = neighbor_model['30'][0].fit(x_rough_scaler, rougher_target)
best_model_clean = neighbor_model['30'][1].fit(x_clean_scaler, cleaner_target)
```

Не забудем отмасштабировать тестовые значения.

```python
def test_predict(model_1, model_2, x_1, y_1, x_2, y_2):
    predict_1 = model_1.predict(x_1)
    predict_2= model_2.predict(x_2)
    
    smape_rought = smape(y_1.values.ravel(), predict_1)
    smape_final = smape(y_2.values.ravel(), predict_2)

    print("First stage sMAPE: {:.2f}%".format(smape_rought))
    print("Final stage sMAPE: {:.2f}%".format(smape_final))

    print('Final sMAPE: {:.2f}%'.format(final_smape(smape_rought, smape_final)))
```

```python
X_first_test_sc = scaler_1.transform(rough_test_input) 
X_second_test_sc = scaler_2.transform(final_test_input)

test_predict(best_model_rought, best_model_clean, 
            X_first_test_sc, rough_test_target,
            X_second_test_sc, final_test_target)
```
