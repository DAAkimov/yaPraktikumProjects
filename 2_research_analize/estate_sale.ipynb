```python
flat_data.loc[:,'balcony'].fillna(0, inplace=True)
flat_data = flat_data.dropna(subset=['floors_total', 'locality_name']).reset_index(drop=True)
```

Для заполнения пропусков в living_area kitchen_area напишем функцию, которая будет рассчитывать необходимые нам параметры. Для этого посчитаем отношение жилой площади и площади кухни к общей и найдем среднее значение для групп зданий по максимальной этажности. Также разделение на этажность может помочь с заполнением пропусков в данных по высоте потолков.

```python
build_dat = flat_data.loc[:,['total_area', 'kitchen_area', 'living_area', 'floors_total', 'ceiling_height']]
build_dat.dropna(inplace=True)
build_dat['kitch_total'] = build_dat['kitchen_area'] / build_dat['total_area']
build_dat['liv_total'] = build_dat['living_area'] / build_dat['total_area']

mean_build_dat = build_dat.pivot_table(index='floors_total',
                                       values=['kitch_total','liv_total', 'ceiling_height'],
                                       aggfunc='median')

mean_build_dat = mean_build_dat.reset_index().set_index('floors_total')
sns.pairplot(mean_build_dat)#.hist(bins=10, figsize=(10,10), setattr);
```

Мы видим, что в столбце "ceiling_height" есть значение, которое в более чем в 3 раза отличается от всех остальных, можем посмотреть, что это за этаж и с чем связан такой разброс.

```python
anomal_floor = mean_build_dat[mean_build_dat['ceiling_height'] == mean_build_dat['ceiling_height'].max()].index[0]
flat_data[flat_data['floors_total'] == anomal_floor]
```

Мы видим, что одно значения равняется 27, это невозможно - скорее всего это просто ошибка при записи данных и настоящее значение равно 2.7

```python
flat_data.iloc[5764, 5] = 2.7
mean_build_dat.iloc[30, 0] = 2.7
```

```python
def add_kitch_liv_area_info(row):
    floor = row['floor']
    total_area = row['total_area']
    kitch_koef = mean_build_dat.iloc[floor]['kitch_total']
    liv_koef = mean_build_dat.iloc[floor]['liv_total']
    
    if (pd.isna(row['kitchen_area']) & (pd.isna(row['kitchen_area']))):
        row['kitchen_area'] = total_area * kitch_koef
        row['living_area'] = total_area * liv_koef
    elif pd.isna(row['kitchen_area']) :
        total_area -= row['living_area']
        row['kitchen_area'] = total_area * kitch_koef
    elif pd.isna(row['living_area']):
        total_area -= row['kitchen_area']
        row['living_area'] = total_area * liv_koef
        
    return row
```

Осталось избавиться от пропусков в высоте потолков

```python
filtr = (flat_data['living_area'].isna()) | (flat_data['kitchen_area'].isna())
flat_data[filtr] = flat_data[filtr].apply(func=add_kitch_liv_area_info, axis=1)
```

```python
def add_ceil_info(row):
    try:
        row['ceiling_height'] = mean_build_dat.loc[row['floors_total']]['ceiling_height']
    except:
        row['ceiling_height'] = 2.75
    return row

filtr = flat_data['ceiling_height'].isna()
flat_data[filtr] = flat_data[filtr].apply(add_ceil_info, axis=1)
```

Добавим классификатор этажа. Напишем функцию, которая делит данные на первый-последний и промежуточный, также создадим группу, в которую будут помещаться значения, в которых мы не уверены.

```python
def floor_rename(row):
    floor = row['floor']
    total_floor = row['floors_total'] if row['floors_total'] != -1 else 0
    if floor == 1:
        return "Первый этаж"
    if floor == total_floor:
        return "Последнйи этаж"
    if floor != total_floor:
        return "Промежуточный этаж"
    return "Нет данных"
```

```python
flat_data = flat_data.reset_index(drop=True)
flat_data['floor_kat'] = flat_data.apply(floor_rename, axis=1)
flat_data['floor_kat_val'] = flat_data['floor_kat'].map({"Первый этаж":0,
                                                         "Последнйи этаж":2,
                                                         "Промежуточный этаж":1,
                                                         "Нет данных":-1})
```

Посчитаем отношение жилой и общей площади, а также отношение площади кухни к общей. Добавим столбцы "liv-tot" и "kitch-tot"

```python
def proporc(row):
    param = row[0]
    total_area = row[1]
    if param == -1:
        return -1
    else:
        return param/total_area
    
    
    
flat_data['liv-tot'] = flat_data[['living_area', 'total_area']].apply(proporc, axis=1)
flat_data['kitch-tot'] = flat_data[['kitchen_area', 'total_area']].apply(proporc, axis=1)
```

Так как большая часть населенных пунктов, присутствующая в выборке содержит мало объявлений, поменяем все название н.п в которых меньше 20 заявок на "Другие"

```python
locality_count = flat_data['locality_name'].value_counts()
locality_mask = flat_data['locality_name'].apply(lambda x: locality_count[x]>=20)
flat_data['locality_name_mask'] = flat_data['locality_name'].where(locality_mask, 'Другие');
```

### 3. Уберем редкие и выбивающиеся значения.
Определеим значения, которые могут понадобиться нам для определения цены квартиры. Построим гистаграммы для численных признаков.


```python
value_column = flat_data.select_dtypes(include=('float64')).columns
not_value_column = flat_data.select_dtypes(include=('bool', 'object', 'datetime64[ns]', 'int64', 'int32')).columns
```

Как и в предыдущем пункте нам мешают выбросы, иpбавимся от них.

Напишем функцию для фильтрации значений по межквартильному размаху, которой пройдемся по всем интересующим нас данным.

```python
def filtr_quantile(columns):
    Q1 = columns.quantile(0.25)
    Q3 = columns.quantile(0.75)
    IQR = Q3 - Q1
    x_min = Q1 - 1.5 * IQR
    x_max = Q3 + 1.5 * IQR
    return (x_min, x_max)
```

```python
flat_data_filtr = pd.DataFrame()
for row in flat_data[value_column]:
    x_min, x_max = filtr_quantile(flat_data[row])
    #print(row,x_min, x_max)
    if row in ['parks_around3000', 'ponds_around3000', 'days_exposition']:
        flat_data_filtr[row] = flat_data[row]
    else:
        filtr = (flat_data[row] >= x_min) & (flat_data[row] <= x_max)
        flat_data_filtr[row] = flat_data[row][filtr]
    

flat_data_filtr = flat_data_filtr.join(flat_data[not_value_column])
```

Мы отфильтровали наши данные на выбросы, посмотрим теерь, нет ли каких-либо аномалий в данных.
